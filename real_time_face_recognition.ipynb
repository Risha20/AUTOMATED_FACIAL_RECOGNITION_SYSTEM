```python
import os
import cv2 as cv
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pickle
from keras_facenet import FaceNet
```

```python
import cv2 as cv
import numpy as np
import pickle
import csv
from sklearn.preprocessing import LabelEncoder
from datetime import datetime

# Initialize models and load embeddings
facenet = FaceNet()
faces_embeddings = np.load("faces_embeddings_done_3.npz")
Y = faces_embeddings['arr_1']

encoder = LabelEncoder()
encoder.fit(Y)
haarcascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')
model = pickle.load(open("SVM_Model_For_Face_Recognition_3.pkl", 'rb'))

# Initialize the webcam (use 0 for the default camera)
cap = cv.VideoCapture(1)

# Prepare attendance record
attendance = {name: "Absent" for name in encoder.classes_}  # Default all as "Absent"

# Create or open CSV file to log attendance
filename = f"attendance_{datetime.now().strftime('%d_%m_%Y')}.csv"
with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Status"])  # CSV Header

    try:
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                print("Failed to capture image")
                break

            rgb_img = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
            gray_img = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
            faces = haarcascade.detectMultiScale(gray_img, 1.3, 9)

            for (x, y, w, h) in faces:
                img = rgb_img[y:y+h, x:x+w]
                img = cv.resize(img, (160, 160))
                img = np.expand_dims(img, axis=0)
                
                # Extract embeddings and predict
                ypred = facenet.embeddings(img)[0]
                face_name = model.predict([ypred])
                final_name = encoder.inverse_transform(face_name)[0]

                # Mark the detected person as "Present"
                attendance[final_name] = "Present"
                
                # Draw rectangle and label on the frame
                cv.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 255), 2)
                cv.putText(frame, final_name, (x, y-10), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv.LINE_AA)

            # Display the frame
            cv.imshow("Face Recognition", frame)
            
            # Check if 'q' is pressed to quit
            if cv.waitKey(1) & 0xFF == ord('q'):
                print("Quitting...")
                break

        # Write the attendance to the CSV file
        for name, status in attendance.items():
            writer.writerow([name, status])

    finally:
        # Release the capture and close windows
        cap.release()
        cv.destroyAllWindows()
        print("Resources released and windows closed.")
```
